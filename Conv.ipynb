```
import numpy as np
from scipy import signal
from scipy import misc
```

```
#1D CONV
x = np.array([0,1,2,3,4])
w = np.array([1,-1,2])

print('x = ',x)
print('w = ',w)

res = np.convolve(x,w)
print('conv(x,w) = ',res)
```

```
x1 = np.array([8,4,2,6,4,10])
w1 = np.array([9,-10,2,-9])

print('x1 = ',x1)
print('w1 = ',w1)

res1 = np.convolve(x1,w1)
print('conv(x1,w1) = ',res1)
```

```
#2D CONV
x = np.array([[1,2,0],[4,1,1],[3,1,2]])
w = np.array([[2,2],[1,-1]])
           
print('x = \n',x)
print('w = \n',w)
             
res1 = signal.convolve2d(x,w,'valid')
res2 = signal.convolve2d(x,w,'same')
res3 = signal.convolve2d(x,w,'full')

print('\n valid conv_2d(x,w) = \n', res1)
print('\n same conv_2d(x,w) = \n', res2)
print('\n full conv_2d(x,w) = \n', res3)
```

```
#EX1
x = np.array([[1,0,2],[1,1,-1],[2,0,2]])
w = np.array([[1,-1],[0,2]])
           
print('x = \n',x)
print('w = \n',w)

res1 = signal.convolve2d(x,w,'valid')
print('\n valid conv_2d(x,w) = \n', res1)
```

```
#EX2
x = np.array([[1,1,2,0],[-1,0,1,1],[2,1,-1,0],[1,0,1,2]])
w = np.array([[0,2],[-1,3]])
           
print('x = \n',x)
print('w = \n',w)

res1 = signal.convolve2d(x,w,'valid')
print('\n valid conv_2d(x,w) = \n', res1)
```
